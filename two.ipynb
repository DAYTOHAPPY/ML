## 案例1 电影分类

### 数据准备

```python
data = {
    '电影名称':['California Man','He is Not Really into Dudes','beautiful Woman','Kevin Longblade','Robo Slayer 3000','Amped 2'],
    '打斗镜头':[3,2,1,101,99,98],
    '接吻镜头':[104,100,81,10,5,2],
    'label':['爱情片','爱情片','爱情片','动作片','动作片','动作片']
}

df= pd.DataFrame(data)
df
```

```python
# 显示df的列切片 索引1-2
df.iloc[:,1:3] 
```

```python
# 求切片中每一行 与 未被分类的数据 的 差  这是为了求 欧式点到点的距离
(df.iloc[:,1:3]-[18,90])**2
```

```python
df1 = (df.iloc[:,1:3]-[18,90])**2
df2 = df1.sum(axis=1)  #  求和 
Distances = pd.DataFrame({'distancesofX':df2**0.5}) # 算出距离 存放在 Distances 中
Distances

```

```python
result = pd.concat([df,Distances],axis=1)  #把df 和 Distances 合并
result
```

```python
result.sort_values(axis=0,by='distancesofX',ascending=True) # 降序排列
```

```python
K=5
result.iloc[0:5,:].value_counts('label') # 取结果
```
